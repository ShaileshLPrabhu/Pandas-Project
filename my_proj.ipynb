<u>Does income and the quantity of products ordered have a relation?</u>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
d= pd.read_csv('customer.csv')
salary= d['Income'].drop_duplicates()
data = {'Income': salary.head(),
        'Quantity': d['Quantity']
       }
df = pd.DataFrame(data,columns=['Income','Quantity'])
# df.plot(x ='Income', y='Quantity', kind = 'line')
sns.barplot(x=df['Income'],y=df['Quantity'],data=df)
plt.xticks(rotation=30,ha='right')
plt.show()
```

Conclusion: In some cases more the income more the quantity but some customers with high income ordered less than they could

<u>Does unit price and quantity have a relation?</u>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
d= pd.read_csv('customer.csv')
data = {'UnitPrice': d['UnitPrice'].tail(10),
        'Quantity':  d['Quantity']
       }
df = pd.DataFrame(data,columns=['UnitPrice','Quantity'])
df.apply(pd.to_numeric)
sns.barplot(x=df['UnitPrice'],y=df['Quantity'])
plt.show()
```

Conclusion:Products with low unit price are ordered in high quantity confirming the assumption.
